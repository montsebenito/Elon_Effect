There are several options to access market data via API using Python. 

In this project, I used `pandas-datareader` library, that provides access to the API endpoints of various data providers through a standard interface.

```python
#Import libraries
#!pip install pandas_datareader
import pandas_datareader as web
import datetime as dt
import pandas as pd
```

```python
#import data since start='2019-01-01'
start = dt.datetime(2019,1,1)
end = dt.datetime.now()
cryptos=['BTC-USD', 'ETH-USD', 'SOL-USD','DOGE-USD', 'SHIB-USD', 'DOT-USD']
for crypto in cryptos:
    pair_split = crypto.split('-')
    data = web.DataReader(crypto, 'yahoo', start, end)
    data['Vol_Fiat'] = data['Volume'] * data['Close'] 
    data.to_csv(f'data/{pair_split[0] + pair_split[1]}_dailydata.csv')
```

```python
#checking data
btc=pd.read_csv('data/BTCUSD_dailydata.csv')
btc.head()
```

```python
#importing other commodity values: Gold, S&P500, Amazon, Tesla and Crude Oil
comm_fin=['GOLD', '^GSPC', 'AMZN','TSLA', 'CL=F']

for symbol in comm_fin:
    data = web.DataReader(symbol, 'yahoo', start, end)
    data['Vol_Fiat'] = data['Volume'] * data['Close'] 
    data.to_csv(f'data/{symbol}_dailydata.csv')
```
